## `scipy.integrate.odeint`

```python
import scipy.integrate as spi
```

```python
f = lambda y,t: np.cos(t)*y
y0 = 1
t = np.linspace(0,10,50)
y = spi.odeint(f,y0,t)
T = np.linspace(0,10,100)
plt.plot(t,y,'.',T,np.exp(np.sin(T)))
plt.show()
```

## Systems of ODEs

$$
ay'' + by' + cy = f(t) \ , \ \ y(0) = y_0 \ , \ \ y'(0) = v_0
$$

```python
N = 100; tf = 2*np.pi;
t = np.linspace(0,tf,N)
h = tf/(N - 1)

a = 1; b = 0; c = 1;
f = lambda t: 0.
y0 = 0; v0 = 1;

y = np.zeros(N)
dy = np.zeros(N)
y[0] = y0
dy[0] = v0

for n in range(0,N-1):
    y[n + 1] = y[n] + dy[n]*h
    dy[n + 1] = dy[n] + (f(t[n]) - b*dy[n] - c*y[n])/a*h

plt.plot(t,y,'.',t,np.sin(t))
plt.show()
```
