## Python Packages

A Python package is a collection of files containing Python code. We import packages with the `import` keyword and access the functions, constants, etc. in the packages using dot syntax.

[NumPy](https://numpy.org/doc/stable/) is the main numerical Python package. It gives us vectors, matrices, constants, and all the usual math functions.

[Matplotlib](https://matplotlib.org) is the main plotting package and it is built on NumPy.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,100)
y = np.cos(np.pi*x) + np.sin(2*np.pi*x)
plt.plot(x,y,'.-')
plt.show()
```

## NumPy Functions

We have all the usual math functions: `np.exp`, `np.cos`, `np.sin`, `np.log`, etc. They operate elementwise on arrays.

```python
x = np.linspace(0,1,5)
y = np.sin(np.pi*x)
y
```

Let's use `np.exp` to investiate the fixed points iteration of $f(x)$.

Does the fixed point iteration for $f(x) = e^{-x}$ converge for $x_0 > 0$? What about $x_0 <= 0$?

```python
def fixed(f,x0,N):
    xn = x0
    for n in range(0,N):
        xn = f(xn)
    return xn
```

```python
fixed(lambda x: np.exp(-x),1,10)
```

## Plotting with Matplotlib

Let's plot the fixed point iteration of $f(x) = e^{-x}$, $x_0 = 1$.

```python
x = np.linspace(-1,2,100)
f = lambda x: np.exp(-x)
y = f(x)
plt.plot(x,y,'b',x,x,'r')
xn = 1
for n in range(0,4):
    fxn = f(xn)
    plt.plot([xn,xn],[xn,fxn],'k.--')
    plt.plot([xn,fxn],[fxn,fxn],'k.--')
    xn = fxn
plt.ylim([-1,3]), plt.grid(True), plt.axis('equal')
plt.show()
```

```python
x = np.linspace(-1.5,1.5,100)
f = lambda x: x*(x-1)*(x+1)
y = f(x)
plt.plot(x,y,'b',x,x,'r')
xn = 0.5
for n in range(0,10):
    fxn = f(xn)
    plt.plot([xn,xn],[xn,fxn],'k.--')
    plt.plot([xn,fxn],[fxn,fxn],'k.--')
    xn = f(xn)
plt.ylim([-1,3]), plt.grid(True), plt.axis('equal')
plt.show()
```
