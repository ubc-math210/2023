* Error formula for left/right Riemann sums
* Example

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Error Formula

$$
E_N^L(f) = \left| \int_a^b f(x)dx - L_N(f) \right| \leq \frac{(b - a)^2}{2N} K_1
$$

## Example

Find $N$ which gurantees $E_N^L(f)$ for $f(x) = \sin(x^2)$ on $[0,\sqrt{\pi/2}]$.

```python
a = 0; b = np.sqrt(np.pi/2)
x = np.linspace(a,b,100)

y = np.sin(x**2)
plt.plot(x,y), plt.title("f(x)")
plt.show()

dydx = 2*x*np.cos(x**2)
plt.plot(x,dydx), plt.title("f'(x)")
plt.show()
```

```python
K1 = np.max(dydx)
K1
```

```python
N =(b - a)**2/(2*0.01)*K1
N
```

```python
N = int(np.ceil(N))
N
```

```python
x = np.linspace(a,b,N+1)
y = np.sin(x**2)
dx = (b - a)/N
I = np.sum(y[:-1])*dx
I
```
