## Trapezoid Rule

Let $f(x)$ be a function defined on $[a,b]$ and let $x_k = a + k \Delta x$ for $k=0,\dots,N$ with $\Delta x = (b-a)/N$. The trapezoid rule is

$$
T_N(f) = \sum_{k=1}^N \left( \frac{f(x_k) + f(x_{k-1})}{2} \right) \Delta x
$$

The error formula for the trapezoid rule is

$$
E_N^T(f) \leq \frac{(b-a)^3}{12N^2}K_2
$$

where $|f''(x)| \leq K_2$ for all $x \in [a,b]$.

## Example

Let's approximate the integral

$$
\int_0^1 \frac{dx}{1 + x^5}
$$

using the trapezoid rule with $N$ which guarantees the error is less than $0.01$. Compute the derivatives of $f(x)$:

$$
f'(x) = \frac{-5x^4}{(1 + x^5)^2}
\hspace{10mm}
f''(x) = -\frac{10 x^3 (2 - 3x^5)}{(1 + x^5)^3}
$$

Plot $f(x)$:

```python
x = np.linspace(0,1,100)
y = 1/(1 + x**5)
plt.plot(x,y)
plt.show()
```

We expect the integral to be between 0.5 and 1. Plot $f''(x)$ to find a bound:

```python
x = np.linspace(0,1,100)
d2y = -10*x**3*(2 - 3*x**5)/(1 + x**5)**3
plt.plot(x,d2y)
plt.show()
```

```python
np.max(np.abs(d2y))
```

```python
a = 0; b = 1; K2 = 3.3;
N = int(np.ceil(np.sqrt((b - a)**3/(12*0.01)*K2)))
N
```

```python
x = np.linspace(a,b,N + 1)
y = 1/(1 + x**5)
dx = (b - a)/N
T = np.sum(y[:-1] + y[1:])/2*dx
T
```
