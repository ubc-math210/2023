## Central Difference Formula

Write a function called `D` which takes `f` and `h` and returns the function

$$
Df = \frac{f(x + h) - f(x - h)}{2h}
$$

```python
def D(f,h=0.01):
    dfdx = lambda x: (f(x + h) - f(x - h))/(2*h)
    return dfdx
```

Check to see if the function returns the expected result:

```python
f = lambda x: np.exp(np.cos(x))
x = np.linspace(0,np.pi,50)
plt.plot(x,f(x))
plt.show()
```

```python
D2f = D(D(f))
d2f = np.exp(np.cos(x))*(np.sin(x)**2 - np.cos(x))
plt.plot(x,D2f(x),x,d2f,'.')
plt.show()
```
